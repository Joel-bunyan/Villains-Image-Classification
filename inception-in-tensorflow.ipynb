```python
import numpy as np
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot  as plt
```

```python
from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale=1.0 / 255)

images = datagen.flow_from_directory('/kaggle/input/villains-image-classification/Villains', target_size=(150,150),
            class_mode='sparse')
```

```python
inception = tf.keras.applications.inception_v3.InceptionV3(
    include_top=False,
    weights='imagenet',
    input_tensor=None,
    input_shape=(150,150,3),
    pooling=None,
    classes=5,
    classifier_activation='softmax'
)
for layer in inception.layers:
    layer.trainable = False
i = tf.keras.layers.Input(shape=(150,150,3))
x1 = inception(i)

x5 = tf.keras.layers.Flatten()(x1)
x6 = tf.keras.layers.Dense(256,activation='relu')(x5)
x7 = tf.keras.layers.Dense(5,activation='softmax')(x6)

model = tf.keras.Model(inputs=(i),outputs=(x7))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(images,epochs=10)
```
